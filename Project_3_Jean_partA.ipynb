Section 1: Data and Research Questions
In this paper, the authors examine the relationship between depth, latitude, and various morphological traits. The aim of the research is to understand how different environmental gradients affect body shape and other key traits. To do this, the authors gathered the following information across 329 sites at different depths ranging from 501200 meters across New Zealand water systems. For each fish, they recorded key features related to functional traits such as total body length, jaw length-to-mouth width ratio, eye size, fin-base to perimeter ratio, pectoral fin-base to body-depth ratio, body transversal shape, and elongation. They also collected general environmental variables like depth and latitude. 



Key details collected: 
* The observations consist of 144 species that were observed across 329 sites at different depths.
* The features are seven morphological traits along with location-based features like depth and latitude.
* Additional data included environmental variables such as light levels and temperature variations, which were inferred or assumed to draw  insight.
* The authors also relied on previous studies to define specialized functional traits. They introduced three functional traits themselves, namely:
* Jaw length-to-mouth width ratio.
* Fin base-to-perimeter ratio.
* Pectoral fin base-to-body depth ratio.

Research Questions
The research question of the authors was: How do depth and latitude influence morphological traits in marine fishes?

Based on prior knowledge suggesting that most fishes in deeper water regions have an enlarged gape to accommodate a wide variety of fishes, they hypothesized that jaw length-to-mouth width ratio would increase with depth.
Additionally, since body length tends to increase with depth, they hypothesized that long fishes that swim using body and caudal-fin undulation (side fins to sort of float) would have a high fin base-to-perimeter ratio, whereas the pectoral fin, which is important for rapid movement, would be smaller at lower depths. This is because rapid movement isnt necessary if all the food is at the bottom of the ocean.
They also hypothesized that eye size would increase with depth due to reduced light.Thus a larger surface area in the eye would be necessary to compensate. 
Finally, they hypothesized that total body length would increase with depth. This is mostly due increase pressure at lower depth, which favors a leaner morphology.

Given that we are interested in variance, a box plot is a good place to start as it can give us a pictorial idea of what to expect. 

```python
!pip install seaborn
!pip install matplotlib

import seaborn as sns
import matplotlib.pyplot as plt

# Box plot for TL
plt.figure(figsize=(8, 6)) # 8 refers to the width, 6 refers to the height
sns.boxplot(data=cleaned_df, y="TL") # we are going to create a box plot with data from cleaned_df, using column TL
plt.title("Box Plot of Total Length (TL)") # title fo the box plt
plt.ylabel("Total Length (TL)") # how we lable the y axis
plt.show()

# Box plot for PBd
plt.figure(figsize=(8, 6))
sns.boxplot(data=cleaned_df, y="PBd")
plt.title("Box Plot of Pectoral Fin Base (PBd)")
plt.ylabel("Pectoral Fin Base (PBd)")
plt.show()

```

TL goes from near zero to 700, showing a much wider range than Pectoral Fin Base. This is suggestive of the diversity of species you are likely to find if you are looking at total body length. The median for TL is between 300 - 400, whereas the median for PBd is 50-100, suggesting there is a tighter clustering of values for PBd. The interquartile range (middle) also spans a wider range for TL, suggesting more diverse body lengths than PBd.
